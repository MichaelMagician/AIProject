# Decision Tree


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
from sklearn.tree import DecisionTreeClassifier
```

```python
def plot_decision_boundary(X, model):
    h = .02
    
    x_min, x_max = X[:, 0].min() -1, X[:,0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    z = z.reshape(xx.shape)
    
    plt.contour(xx, yy, z, cmap=plt.cm.Paired)
```

### Read data

```python
np.random.seed(1)
N = 500
D = 2
X = np.random.randn(N,D)

delta = 1.5
X.shape
X[: N//2, :] += np.array([delta, delta])
X[ N//2:, :] += np.array([-delta, -delta])

Y = np.array([0] * (N//2) + [1] * (N//2))
plt.scatter(X[:,0], X[:,1], s=100, c=Y, alpha=0.5)
```

### Train model

```python
model = DecisionTreeClassifier()
model.fit(X, Y)
print(f'score for basic model {model.score(X, Y)}')

plt.scatter(X[:,0], X[:,1], s=100, c=Y, alpha=0.5)
plot_decision_boundary(X, model)
plt.show()
```

```python

model_5 = DecisionTreeClassifier(criterion='entropy',max_depth=5)
model_5.fit(X, Y)
print(f'score for basic model {model_5.score(X, Y)}')

plt.scatter(X[:,0], X[:,1], s=100, c=Y, alpha=0.5)
plot_decision_boundary(X, model_5)
plt.show()
```

```python
model_3 = DecisionTreeClassifier(criterion='entropy',max_depth=3)
model_3.fit(X, Y)
print(f'score for basic model {model_3.score(X, Y)}')

plt.scatter(X[:,0], X[:,1], s=100, c=Y, alpha=0.5)
plot_decision_boundary(X, model_3)
plt.show()
```
